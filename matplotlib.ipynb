```python
import matplotlib.pyplot as mp
import numpy as np
```

```python
l=["s1","s2","s3","s4","s5","s6","s7","s8"]
l1=[35,80,32,72,76,20,99,55]
```

```python
mp.plot(l,l1,color="red" , marker="h",mec="Blue",)
mp.title("Student Data")
mp.xlabel("Student_Name")
mp.ylabel("Marks")
mp.show()
```

```python
mp.bar(l,l1)
mp.title("Student Data")
mp.xlabel("Student_Name")
mp.ylabel("Marks")
mp.show()
```

```python
mp.barh(l,l1,color="Red")
mp.title("Student Data")
mp.xlabel("Student_Name")
mp.ylabel("Marks")
mp.show()
```

```python
mp.scatter(l,l1)
mp.title("Student Data")
mp.xlabel("Student_Name")
mp.ylabel("Marks")
mp.show()
```

```python
x=[34,30,56,69,89,43,78,12,45]
mp.pie(x)
mp.title("pie chart")
mp.show()
```

```python
x=[34,30,56,69,89,43,78,12,45]
mp.hist(x)
mp.show()
```

```python
mp.step(l,l1)
mp.title("Student Data")
mp.xlabel("Student_Name")
mp.ylabel("Marks")
mp.show()
```

```python
mp.stem(l,l1)
mp.title("Student Data")
mp.xlabel("Student_Name")
mp.ylabel("Marks")
mp.show()
```
